```python
def setvisible(self,vis):
    for c in self.collections: c.set_visible(vis)
def setanimated(self,ani):
    for c in self.collections: c.set_animated(ani)
```

```python
def Te_anim():
   # fig,ax = plt.subplots(figsize=(10,10))
   # ax.set_xlabel("R (m)")
   # ax.set_ylabel("Z (m)")
   # ax.contour(pxr.X[:,:,0]*a,pxr.Z[:,:,0]*a,psi_pol[:,:,0,0],20,colors="k")
   # ax.set_aspect("equal")
   # fig = plt.figure()
    ims = []
    for timestamp in range(0,3):
        im1 = plt.contourf(pxr.X[:,:,0]*a,pxr.Z[:,:,0]*a,Te[:,:,0,timestamp]*Tmax,40,cmap="coolwarm")
        text = 't = '+str(timestamp)
        an = plt.annotate(text, xy=(0.8, 0.94), xycoords='axes fraction',fontsize=14)
        im1.set_visible = types.MethodType(setvisible,im1)
        im1.set_animated = types.MethodType(setanimated,im1)
        
        im1.axes = plt.gca()
        im1.figure = fig
        ims.append(im1.collections)
        #cbar = ax.colorbar()
        #cbar.set_label(r"$T_e$ (keV)")
    art_ani = animation.ArtistAnimation(fig, ims, interval=200, repeat_delay=100,blit=False)
    return art_ani
```
