```julia
Psi_int = Interpolations.interpolate(Psi_un2,(BSpline(Quadratic(Line(OnGrid()))),BSpline(Quadratic(Periodic(OnGrid()))),BSpline(Quadratic(Periodic(OnGrid())))));
r = range(0.0, length=size(Psi_un2)[1], stop=1.0);
u = range(0.0, length=size(Psi_un2)[2], stop=2.0*pi);
fi = range(0.0, length=size(Psi_un2)[3], stop=2.0*pi);
Psi_sint = scale(Psi_int, r,u,fi);
Psi_eint = extrapolate(Psi_sint, Line());
```

```julia
function psi_proj()
    psi_list = []
    uf_list = []
    psi_int_list = []
    ls = []
    psi_min = -0.0030128203568753017
    norm = 0.5341460400872815
    for i in keys(dict1)
        l_min = dict1[i[1],"0"]["l_u"](0.001)
        l_max = dict1[i[1],"0"]["l_u"](6.28)
        ls = range(l_min,length=30,stop=l_max)
        for l in ls
            r = dict1[i[1],"0"]["r_l"](l)
            u = dict1[i[1],"0"]["u_l"](l)
            uf = dict1[i[1],"0"]["uf_l"](l)
            append!(psi_list,(parse(Float64,i[1])-psi_min)/(norm-psi_min))
            append!(uf_list,uf)
            append!(psi_int_list,(Psi_eint(r,u,0.0)-psi_min)/(norm-psi_min))
        end            
    end
    spl = Spline2D(uf_list,psi_list,psi_int_list,s=0.0001)
    #nodes = ([psi_list],[uf_list])
    #itp = Interpolations.interpolate(nodes, psi_int_list, Gridded(Linear()))
    #triin = Triangulate.TriangulateIO()
    #println(uf_list)
    #temp = hcat(psi_list,uf_list)
    #triin.pointlist = permutedims(temp,(2,1))
    #println(size(triin.pointlist,1))
    #(triout,vorout) = triangulate("Q",triin)
    #uf_list_r = reverse(uf_list)
    #psi_list_r = reverse(psi_list)
    #ExtendableGrids.simplexgrid(uf_list_r,psi_list)
    #x = [0.5,0.5]
    #y = GridInterpolations.interpolate(triout,psi_int_list,x)
    #println(y)
    return spl
end
```
