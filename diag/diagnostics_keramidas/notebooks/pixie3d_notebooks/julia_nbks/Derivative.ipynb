```julia
function dAdr(arr,theta,fi,t)
    dr = 1.0 / size(arr,1)
    psi = arr[:,theta,fi,t]
    psiplus = circshift(psi,(-1))
    psiminus = circshift(psi,(1))
    dpsidr = (psiplus - psiminus)/(2.0*dr)
    dpsidr_start = (psi[2]-psi[1])/dr
    dpsidr_end = (psi[length(psi)]-psi[length(psi)-1])/dr
    dpsidr[1] = dpsidr_start
    dpsidr[length(psi)] = dpsidr_end
    return dpsidr
end
```

```julia
function deriv(arr)
    t_deriv = [] 
    for theta in range(1,stop=size(arr,2))
        for fi in range(1,stop=size(arr,3))
            t_deriv = append!(t_deriv,dAdr(arr,theta,fi,1))
        end
    end
    return permutedims(reshape(t_deriv,(size(arr,1),size(arr,3),size(arr,2))),(1,3,2))
end
```

```julia
function dAdt(arr,r,fi,t)
    dtheta = 2.0*pi / size(arr,2)
    psi = arr[r,:,fi,t]
    psiplus = circshift(psi,(-1))
    psiminus = circshift(psi,(1))
    dpsidtheta = (psiplus - psiminus)/(2.0*dtheta)
    dpsidtheta_start = (psi[2]-psi[1])/dtheta
    dpsidtheta_end = (psi[length(psi)]-psi[length(psi)-1])/dtheta
    dpsidtheta[1] = dpsidtheta_start
    dpsidtheta[length(psi)] = dpsidtheta_end
    return dpsidtheta
end
```

```julia
function deriv_t(arr)
    t_deriv = [] 
    for r in range(1,stop=size(arr,1))
        for fi in range(1,stop=size(arr,3))
            t_deriv = append!(t_deriv,dAdt(arr,r,fi,1))
        end
    end
    return permutedims(reshape(t_deriv,(size(arr,2),size(arr,3),size(arr,1))),(3,1,2))
end
```
